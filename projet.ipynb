```python
import pandas as pd

df = pd.read_csv("store.csv")
df
```

```python
df.columns
```

```python
df.columns = (df.columns.str.replace(" ", "_"))
df.columns
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset="Postal_Code")
```

```python
df.isnull().sum()
```

```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"],format="%d/%m/%Y")
df["Ship_Date"] = pd.to_datetime(df["Ship_Date"],format="%d/%m/%Y")

df.dtypes
```

```python
df.drop_duplicates()
```

```python
df["Postal_Code"]=df["Postal_Code"].astype(int)
df.dtypes
```

```python
df["Année"]= df["Ship_Date"].dt.year
df["Mois"]= df["Ship_Date"].dt.month
df["Trimestre"]= df["Ship_Date"].dt.quarter
df
```

```python
df["Cost"] = df["Sales"] * 0.6
df
```

```python
df["marge"] = df["Sales"] - df["Cost"]
df["marge"].round(2)
```

```python
df["taux_de_profit"] = df["marge"] / df["Sales"] * 100
df["taux_de_profit"]
```

```python
def seg_clients(valeur):
    if valeur < 500:
        return "High"
    elif valeur < 250:
        return "Medium"
    else:
        return "Low"
df["segmentation_clients"]= df["Sales"].apply(seg_clients)
df
```

```python
df.columns
```

```python
df["délais_livraison"] = df["Ship_Date"] - df["Order_Date"]
df
```

```python
def Performance(day):
    if day.days <= 3:
        return "in time"
    else:
        return "late"

df["performance"] = df["délais_livraison"].apply(Performance)
df
```

```python
df["Region"].unique()
```

```python
df["Product_Name"].unique()
```

```python
product_sales = df.groupby("Product_Name")["Sales"].sum()
product_sales
```

```python
Region_sales = df.groupby("Region")["Sales"].sum().round(2)
Region_sales
```
